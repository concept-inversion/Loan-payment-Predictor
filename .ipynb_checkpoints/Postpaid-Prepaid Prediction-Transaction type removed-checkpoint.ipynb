```python
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

```python
df = pd.read_csv("DATA-POSPAY.csv",sep=';')
```

Fill missing datapoints with 0

```python
df= df.fillna(0)
```

Translate the headers

```python
columns_translation = {
    "idtransaksi":"type_of_transaction",
    "noloket": "counter_id",
    "nopen": "branch_id",
    "tanggal": "date_of_transaction",
    "jam": "hour",
    "wilpos": "high_level_territory",
    "idpetugas": "staff_id",
    "kdmitra": "product_id",
    "besaruang": "payment_amount",
    "detail": "detail",
    "feemitra": "PTPOS_fee"
}
```

```python
df = df.rename(index=str, columns=columns_translation)
```

```python
df
```

Removing the comma from the payment amount column

```python
df["payment_amount"] = [x.split(',')[0] for x in df['payment_amount'].astype(str)]
```

The detail column has many different formats for the data in it.  
We're concerned with the data for the prepaid and postpaid columns only, whose formats we've predetermined.  
Postpaid data has 38 sub-columns in the details column whereas Prepaid data has 28 sub-columns.  
  
Calculate the number of sub-columns in the details column by splitting it on '~' and append it to each row.

```python
df["detail2"]= [len(x.split('~'))for x in df['detail'].astype(str)]
```

Get a dataframe containing just the Postpaid data. Drop the sub-column count column.

```python
df2 = df.loc[df["detail2"]==38].reset_index(drop=True).drop(["detail2"],axis=1)
```

From the known format of the "detail" column, extract the PLN Electricity group and the customer ID. Add them both as new columns.

```python
df2["PLN_eletricity_group"] = [x.split('~')[6].strip() for x in df2['detail'].astype(str)]
```

```python
df2["customer_ID"] = [x.split('~')[2].strip() for x in df2['detail'].astype(str)]
```

```python
len(df2)
```

Repeat fot the Prepaid data.

```python
df3 = df.loc[df["detail2"]==28].reset_index(drop=True).drop(["detail2"],axis=1)
```

```python
df3["PLN_eletricity_group"] = [x.split('~')[10].strip() for x in df3['detail'].astype(str)]
```

```python
df3["customer_ID"] = [x.split('~')[1].strip() for x in df3['detail'].astype(str)]
```

```python
len(df3)
```

Merge the prepaid and postpaid data into a one.

```python
final_df  = pd.concat([df2[:150000],df3[:150000]]).reset_index(drop=True)
```

```python
final_df
```

Drop the "detail" field as we dont need it anymore.

```python
final_df = final_df.drop(["detail"],axis=1)
```

Clear memory

```python
df2 = None
df3 = None
```

```python
def extract_data(df):
    # Extract label from dataframe
    y_data = df["product_id"]

    y_data = y_data.astype("category")
    y_data = y_data.cat.codes

    x_data = df.copy(True)

    x_data = x_data.drop(["product_id","date_of_transaction","hour",
                          "PTPOS_fee","payment_amount","type_of_transaction"],axis=1)
    
    # Columns dropped
    # product_id is the label
    # date_of_transaction and hour, not relevant
    # payment_amount, PTPOS_fee give away the label

    for column in x_data:
        x_data[column] = x_data[column].astype('category')
        x_data[column] = x_data[column].cat.codes
    return x_data,y_data
```

```python
x_data,y_data = extract_data(final_df)
```

```python
x_data.head()
```

```python
def RFC(X_train,X_test, y_train ,y_test):
    clf = RandomForestClassifier(n_estimators = 1000,
                                 max_features='sqrt',
                                 max_depth=None,
                                 n_jobs=-1)
    clf.fit(X_train, y_train.ravel())
    score = clf.score(X_test,y_test.ravel())
    return (clf,score)
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=42)
```

```python
clf,score = RFC(X_train,X_test, y_train ,y_test)
```

```python
score
```

```python
def plot_importances(clf,x_data):
    importances = clf.feature_importances_
    std = np.std([tree.feature_importances_ for tree in clf.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]
    labels=[]
    for x in indices:
        labels.append(x_data.columns[x])
    X=x_data

    plt.figure(figsize=(30,5))
    plt.title("Feature importances")
    plt.bar(range(X.shape[1]), importances[indices],
           color="r", yerr=std[indices],align="center")
    plt.xticks(range(X.shape[1]), labels)
    plt.xlim([-1, X.shape[1]])
    plt.show()
```

```python
plot_importances(clf,x_data)
```
